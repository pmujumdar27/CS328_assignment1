```python
columns = columns[columns < '2010']
columns
```

```python
energy_df
```

```python
energy_df = energy_df.bfill(axis='columns')
energy_df = energy_df.ffill(axis='rows')
```

```python
common = energy_df.index.intersection(gdp_df.index).intersection(pop_df.index)
common = list(common)
# print(common)
common.remove('Kuwait')
common.remove('United Arab Emirates')
common.remove('Norway')
common.remove('Saudi Arabia')
common.remove('Singapore')
common.remove('Switzerland')
common.remove('Qatar')

gdp_df = gdp_df.loc[common]
pop_df = pop_df.loc[common]
energy_df = energy_df.loc[common]
```

```python
melted_energy = energy_df.reset_index().melt('country')
melted_energy.rename(columns={'value':'total_energy_production'}, inplace=True)
melted_energy
```

```python
melted_gdp = gdp_df.reset_index().melt('country')
melted_gdp.rename(columns={'value':'gdp_per_capita'}, inplace=True)
melted_gdp
```

```python
melted_pop = pop_df.reset_index().melt('country')
melted_pop.rename(columns={'value':'population'}, inplace=True)
melted_pop
```

```python
agg_df = melted_energy.copy()
agg_df['population'] = melted_pop['population']
agg_df['gdp_per_capita'] = melted_gdp['gdp_per_capita']
```

```python
agg_df.rename(columns={'variable':'year'}, inplace=True)
agg_df
```

```python
fig = px.scatter(agg_df, x='gdp_per_capita', y='total_energy_production', color='country', size='population', size_max=80, animation_frame='year', range_x=[100,65000], range_y=[-100000,1800000], title="Total Energy production vs GDP per capita", log_x=True)
fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 800
fig.layout.updatemenus[0].buttons[0].args[1]["transition"]["duration"] = 800
```

```python
fig.write_html('total_energy_gdp_pop.html')
```
